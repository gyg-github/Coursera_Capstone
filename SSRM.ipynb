### Description of the problem, discussion and target audience

### Target audience 
is local entrepreneurs and real estate agents.

### Description of the problem
Small shops are sensitive to their closeness to customers, transportation, etc.
Unless it is in the place where a big number of similar small shops exists, it is not clear where new shops should be opened.

### Discussion
Some shops tend to concentrate in one place like style boutiques (self attracting), 
or cover the area uniformly like convenience stores (self repelling), depending on type of shop.
We classify pattern of self/co co-location per business type based on (co)location of shops of possibly different types.
Then we rank areas based on patterns and make top 5 suggestions for each type of shop.

Possible target audience is local entrepreneurs and real estate agents


## What data will be used, how, and examples of data
We use Manhattan FourSquare information to build classification of self/co repelling of the shop types; 
### What data wilto used
We will use ionformation about neighborhood from Manhattan, that has a total of 5 boroughs and 306 neighborhoods from NYU
and detailed information about the neighborhoods available from Foursquare.
